## x17 ’justices’

```python
def extract_justice_info(justice_list):
    # Extract the political direction of each justice in the list
    political_directions = [j['political_direction'] for j in justice_list]
    
    # Return a Series with counts of liberal and conservative justices, and the total number of justices
    return pd.Series({
        'liberal_justices': political_directions.count('Liberal'),
        'conservative_justices': political_directions.count('Conservative'),
        'total_justices': len(political_directions)
    })

# Apply the extract_justice_info function to the 'justices' column of the training data
justice_info = train_data['justices'].apply(extract_justice_info)

# Display the first few rows of the resulting DataFrame
justice_info.head()
```
